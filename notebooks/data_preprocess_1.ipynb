# Titanic Database

```python
df_1 = sns.load_dataset('titanic').drop(columns='alive')
```

```python
df_2 = pd.read_csv('../raw_data/ML_Titanic_dataset.csv')
```

```python
df_2
```

# Preprocess Data

## drop null values

```python
df_1.isnull().sum()
```

```python
df_2.isnull().sum()
```

```python
df_1 = df_1.drop(columns = 'deck')
```

```python
df_1
```

```python
df_2 = df_2.drop(columns = ['Cabin'])
```

```python
df_2
```

## extract titles from names

```python
df_2['title'] = df_2.Name.str.extract(' ([A-Za-z]+)\.',expand=False)
```

```python
df_2
```

```python
df_2['title']= df_2['title'].replace(['Capt','Col','Countess','Don','Dr',\
                           'Jonkheer','Lady','Major','Rev','Sir'],'Rare')
df_2['title']= df_2['title'].replace('Mlle','Miss')
df_2['title']= df_2['title'].replace('Ms','Miss')
df_2['title']= df_2['title'].replace('Mme','Mrs')
```

```python
title_df = df_2.groupby('title').agg({'PassengerId':'count','Survived':'sum'})
title_df
```

```python
title_df.plot(kind='barh',stacked=True)
```

## aggregate family size

```python
df_2['family size'] = df_2['SibSp']+ df_2['Parch']
df_2
```

```python
fam_df = df_2.groupby('family size').agg({'PassengerId':'count','Survived':'sum'})
fam_df
```

```python
fam_df.plot(kind='barh',stacked=True)
```

```python
df_2 = df_2[['title','family size']]
```

```python
df_2
```

## concat two dataframes

```python
df = pd.concat([df_1,df_2],axis=1)
```
